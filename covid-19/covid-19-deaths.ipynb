```python
import pandas as pd
import json

PATH = './COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/'

with open('./COVID-19-master/csse_covid_19_data/country_convert.json', 'r', encoding='utf-8-sig') as json_file:
    json_data = json.load(json_file)

def change_country_region(row):
    if row['Country_Region'] in json_data:
        row['Country_Region'] = json_data[row['Country_Region']]
    return row['Country_Region']

def create_data_frame(file):
    doc = pd.read_csv(PATH + file, encoding='utf-8-sig')

    try:
        doc = doc[['Country_Region', 'Deaths']]
    except:
        doc = doc[['Country/Region', 'Deaths']]
        doc.columns = ['Country_Region', 'Deaths']

    doc = doc.dropna(subset = ['Deaths'])
    doc = doc.dropna(subset = ['Country_Region'])

    doc = doc.astype({'Deaths': 'int64'})
    doc['Country_Region'] = doc.apply(change_country_region, axis=1)

    doc = doc.groupby('Country_Region').sum()

    date_column = file.split('.')[0].lstrip('0').replace('-','/')

    doc.columns = [date_column]
    return doc
```

```python
# 날짜로 변환
import os
import pandas as pd
from datetime import datetime

def generate_dataframe_by_path(PATH):
    file_list = os.listdir(PATH)
    csv_files = []
    first = True
    for file in file_list:
        if file.split('.')[-1] == 'csv':
            csv_files.append(file)
    csv_files.sort(key=lambda x: datetime.strptime(x, '%m-%d-%Y.csv'))

    for csv_file in csv_files:
        doc = create_data_frame(csv_file)
        if first:
            final_doc = doc
            first = False
        else:
            final_doc = pd.merge(final_doc, doc, how='outer', left_index=True, right_index=True)

    final_doc = final_doc.fillna(0)
    return final_doc
```

```python
PATH = './COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/'
doc = generate_dataframe_by_path(PATH)
```
